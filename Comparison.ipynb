```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the stock price data
stock_data = pd.read_csv('stock_data.csv')

# Load the financial statement data (balance sheet, cash flow statement, income statement)
balance_sheet_data = pd.read_csv('balance_sheet.csv')
cash_flow_data = pd.read_csv('cash_flow_statement.csv')
income_statement_data = pd.read_csv('income_statement.csv')
```

```python
# Extract the relevant financial data for comparison
net_income = pd.to_numeric(income_statement_data.iloc[4, 2:], errors='coerce')
total_assets = pd.to_numeric(balance_sheet_data.iloc[0, 2:], errors='coerce')
total_equity = pd.to_numeric(balance_sheet_data.iloc[2, 2:], errors='coerce')

# Extract the relevant stock price data for comparison
stock_prices = stock_data['Close']

# Normalize the financial data and stock prices
normalized_net_income = net_income / net_income.iloc[0]
normalized_total_assets = total_assets / total_assets.iloc[0]
normalized_total_equity = total_equity / total_equity.iloc[0]
normalized_stock_prices = stock_prices / stock_prices.iloc[0]
```

```python
# Plot the normalized financial data and stock prices
plt.figure(figsize=(10, 6))
plt.plot(net_income.index, normalized_net_income, marker='o', label='Net Income')
plt.plot(total_assets.index, normalized_total_assets, marker='o', label='Total Assets')
plt.plot(total_equity.index, normalized_total_equity, marker='o', label='Total Equity')
plt.plot(stock_prices.index, normalized_stock_prices, marker='o', label='Stock Prices')
plt.xlabel('Date')
plt.ylabel('Normalized Value')
plt.title('Comparison of Financial Statements and Stock Prices')
plt.legend()
plt.show()
```
